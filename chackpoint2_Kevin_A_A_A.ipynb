##**Exploratory Data Analysis**

```python
data_visualization = df[['rank', 'personName', 'country', 'finalWorth']]

data_visualization = data_visualization.sort_values(by='rank')

data_visualization = data_visualization.head(20)

plt.figure(figsize=(10, 8))
bars = plt.barh(data_visualization['personName'], data_visualization['finalWorth'], color='skyblue')

for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f"${bar.get_width():,.0f}",
             va='center', ha='left', fontsize=10)

plt.xlabel('Kekayaan Bersih (Dolar)', fontsize=12)
plt.ylabel('Nama Orang', fontsize=12)
plt.title('Top 20 kekayaan bersih Orang Terkaya di Dunia', fontsize=14)
plt.gca().invert_yaxis()

plt.gca().xaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))

plt.show()
```

Ini merupakan visualisai yang menampilkan Top 20 kekayaan bersi orang terkaya didunia. Bisa kita liat bahwa rank dan juga kekayaan bersih yang tertinggi itu di pegang oleh Barnard Arnault & Family yang mana total kekayaan bersinya mencapai 211.000 Bilion Dolar Sedang kan yang terbawah di pegang oleh Rob Walton & Family yang memiliki total kekyaan sekitar 57,600 Bilion Dolar

```python
plt.figure(figsize=(30, 5))

industry_counts = df['industries'].value_counts()

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=industry_counts.values, y=industry_counts.index, palette='colorblind')

for p in ax.patches:
    width = p.get_width()
    plt.text(width, p.get_y() + p.get_height() / 2, f'{int(width)}', va='center')

plt.xlabel('Jumlah')
plt.title('Jumlah bisnis Industri yang para miliarder jalankan')

plt.show()
```

Ini merupakan data visualisasi dari jumlah bisnis industri yang di jalankan para orang kaya di dunia. industri yang banyak di jalankan oleh para orang-orang terkaya ini yaitu bergerak dalam bidang Finance & Investmants yang mana jumlahnya sekitaran 372 industri yang di jalanakan dan sedangkan yang terendah sekitaran 25 industri yaitu di bidang Gambling & Casinos.
